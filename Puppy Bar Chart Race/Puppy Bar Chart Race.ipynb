```python
import bar_chart_race as bcr   #make sure you installed the package using `image_labels` branch.
import pandas as pd
```

```python
#read data
pups_raw = pd.read_csv("puppy_weights.csv").set_index('Date')
```

```python
puppy_headers = ['Emma','Simba','Ragnar','Luna','Big Black','Black1','Black2','Tiny','Jucumari','Rolo']
```

```python
pups = pups_raw[puppy_headers]
pups.head(20)
```

```python
pups.plot()
```

```python
pups.interpolate(axis=0,inplace=True)
pups.tail(30)
```

```python
bcr.bar_chart_race(pups.head(30),
                  img_label_folder='bar_image_labels',
                  title='Puppy Weights (g)',
                  steps_per_period=50,
                  tick_image_mode='trailing',
                  fig_kwargs={'dpi':150},
                  period_label={'x': .99, 'y': .05, 'ha': 'right', 'va': 'center'},
                  orientation='h')
```
